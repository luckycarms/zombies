```python
# Import necessary libraries

import pandas as pd 
from bs4 import BeautifulSoup
import requests
```

```python
# followed https://www.youtube.com/watch?v=8dTpNajxaH0

url = 'https://walkingdead.fandom.com/wiki/The_Walking_Dead_(TV_Series)'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

```

```python
# check to see if it is pulling in webpage
print(soup)
```

```python
table = soup.find_all('table', {'class': 'wikitable'})[1]

print(table)
```

```python
# Getting headers

headers = table.find_all('th')

print(headers)

```

```python
# remove \n
for header in headers:
  
    header_text = header.text.strip()

    # If 'The Walking Dead' is in the header text, remove it from the table
    if 'The Walking Dead' in header_text:
        header.decompose()  # This removes the header element from the tree

# Create a list of headers without 'The Walking Dead'
header_table = [header.text.strip() for header in table.find_all('th')]

# Print the modified list of headers
print(header_table)
```

```python
# create a dataframe
zombies_twd = pd.DataFrame(columns= header_table)
```

```python
zombies_twd
```

```python
rows = table.find_all('tr')
print(rows)
```

```python
# create a loop to imput data from web and then input next set of data starting from position with data
for row in rows[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)

    length = len(zombies_twd)
    zombies_twd.loc[length] = individual_row_data
```

```python
zombies_twd
```

```python
url2 = 'https://walkingdead.fandom.com/wiki/Fear_the_Walking_Dead'

page2 = requests.get(url2)

soup2 = BeautifulSoup(page2.text, 'html')
```

```python
table2 = soup2.find_all('table', {'class': 'wikitable'})[1]

print(table2)
```

```python
# Getting headers

headers2 = table2.find_all('th')

print(headers2)
```

```python
# remove \n
for header in headers2:
  
    header_text2 = header.text.strip()

    # If 'Fear The Walking Dead' is in the header text, remove it from the table
    if 'Fear the Walking Dead' in header_text2:
        header.decompose()  

# Create a list of headers without 'Fear The Walking Dead'
header_table2 = [header.text.strip() for header in table.find_all('th')]

# Print the modified list of headers
print(header_table2)
```

```python
# create a dataframe
zombies_fear = pd.DataFrame(columns= header_table2)
```

```python
rows2 = table2.find_all('tr')
print(rows2)
```

```python
# create a loop to imput data from web and then input next set of data starting from position with data
for row in rows2[2:]:
    row_data2 = row.find_all('td')
    individual_row_data2 = [data.text.strip() for data in row_data2]
    print(individual_row_data2)

    length = len(zombies_fear)
    zombies_fear.loc[length] = individual_row_data2
```

```python
# Stacking the dataframes one on top of the other
zombies_twd_fear = pd.concat([zombies_twd, zombies_fear])
```
